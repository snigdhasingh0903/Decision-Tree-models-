```python
import pandas as pd
import numpy as np
import sklearn 
```

```python
my_file="https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/Exercise/titanic.csv"
df1=pd.read_csv(my_file)
df1.head()
```

```python
df2=df1.drop(["PassengerId","Name","SibSp","Parch","Ticket","Cabin","Embarked"],axis='columns')
df2.head()
df2.dropna()
target=df2['Survived']
input=df2.drop(['Survived'],axis='columns')
input.Sex = input.Sex.map({'male': 1, 'female': 2})
input.Age = input.Age.fillna(input.Age.mean())
input.head()
```

```python
from sklearn.model_selection import train_test_split

input_train,input_test,target_train,target_test=train_test_split(input,target,test_size=0.2)
input_train.head(15)
```

```python
df2.head(15)
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(input_train,target_train)
```

```python
model.score(input_test,target_test)
```
