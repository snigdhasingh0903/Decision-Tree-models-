```python
import pandas as pd
import numpy as np
import matplotlib
```

```python
df1=pd.read_csv("./movie_success_rate.csv")
df1.head()
```

```python
df2=df1['Rating']
df3=df1.dropna()
df3.isnull().sum()
```

```python

d=df3['Year'].unique()
len(d)
```

```python
df4=df3.drop(['Title','Director','Genre','Description','Actors'],axis='columns')
df4.head()
```

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
le_year=LabelEncoder()
```

```python
input=df4.drop(['Success'],axis='columns')
target=df4['Success']
input['year_n']=le_year.fit_transform(input['Year'])
input=input.drop(['Year'],axis='columns')
input.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input,target,test_size=0.2)
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
